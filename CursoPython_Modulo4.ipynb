# Writing Files with Open

Si utilizamos la función de open() con un nombre de archivo que no existe -y modo w-, se crea el archivo. Podemos hacer lo siguiente:

```
with open("/resources/data/Example2.txt", ""w) as file1:
  file1.write("This is line A")
  file1.write("This is line B")
```

Esto creará un nuevo archivo y escribirá dos líneas en él.

Asimismo, si tenemos una lista con varios datos, podríamos cargarlos a un archivo así:

```
#lista de datos:
lines = ["dataA", "dataB", "dataC"]

#inserción:
with open("/resources/data/Example2.txt", "w") as file1:
  for line in lines
    file1.write(line)
```

Si quisieras agregar lineas o datos a un archivo con información existente, podríamos usar el modo "a", en lugar de "w".

En ese caso, solo estamos haciendo *append*.

---

Por último, para **copiar** un archivo a otro nuevo, podemos hacer lo siguiente:

```
with open("archivo1", "r") as copyFile
  with open("archivo2, "w"") as pasteFile
    
    for line in copyFile
      pasteFile.write(line)
```
