Tuplas

* tuple = ('disco', 10, 1.1)
* tuple[0]
* tuple[0:3] -> del subínidce 0 al 2
* len()
* si dos tuplas son iguales, referencian la misma tupla
* las tuplas son INMUTABLES
* sortedTuple = sorted(tuple)
* podemos manipularlo como un arreglo si tenemos nested tuples -> tuple[2][1]

Listas
* CORCHETES, no paréntesis
* son MUTABLES
* se acceden como las tuplas

Podemos extenderlas
* lista.extend([1, 2]) -> agrega dos elementos
* lista.append([1, 2]) -> agrega UN elemento -> inserta una lista dentro de una lista

O cambiar elementos
* listas[0]= otroElemento
* o elimnar con del(lista[0])

O transformar string a list
* 'hola mundo'.split() -> ['hola', 'mundo']
* 'A,B,C'.split(',') -> delimitamos por coma

Para hacer una copia de listas:
* listaB = listaA[:]
* porque si usamos listaB = listaA, estamos apuntando al mismo lugar de memoria. Y si cambiamos "uno", cambiamos al "otro"










```python
lista = [0, 1, 2]
print(lista)
lista2 = lista.extend([3, 4])
print(lista2)
```

```python
a=[0,1,2]
print(a[1:])
```
