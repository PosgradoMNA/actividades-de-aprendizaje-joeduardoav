**Loops**
* range(3) -> [0, 1, 2]
* range(10, 13) -> [10, 11, 12]

```python
squares=['red', 'blue', 'yellow']
for i in range(3):
  squares[i]='white'
print(squares)
print('\n')
####
####
####
for square in squares:
  print(square+'\n')
####
####
####
print('\n')
for i, square in enumerate(squares):
  print(str(i)+': '+square)
```

**While loops**

```python
squares=['red', 'red', 'blue', 'red']
newSquares=[]

i=0
while(squares[i]=='red'):
  newSquares.append(squares[i])
  i=i+1

print(newSquares)
```

**Functions**
* len(lista)
* sum(lista)
* sorted(lista) -> genera nueva lista
* lista.sort() -> no genera nueva lista

```python
def add1(a, b):
  r = a + b
  return r

def add2(*nums):
  r = 0
  for num in nums:
    r = r + num
  return r

def add3(*nums):
  return sum(nums)

suma = add1(2,4)
print(suma)

sumaInteligente = add3(1,2,3,4,5)
print(sumaInteligente)
#wow
```

**Class Circle**


```python
class Circle(object):
  def _init_(self, radius, color):
    self.radius = radius;
    self.color = color;

  def add(self, r):
    self.radius = self.radius + r
    return(self.radius)

redCircle=Circle(10, "red")
print(redCircle.color)
```
